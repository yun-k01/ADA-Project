```python
import pandas as pd
from scipy.stats import mannwhitneyu


superhost = df.loc[df['host_is_superhost'] == 1]
non_superhost = df.loc[df['host_is_superhost'] == 0]

# Loop through columns and perform Mann-Whitney U test
for col in df.columns:
    if col == 'host_is_superhost':
        continue
    p_value = mannwhitneyu(superhost[col], non_superhost[col]).pvalue
    if p_value < 0.05:
        print(f"Column {col} is significantly different between Superhost and non-Superhost (p-value = {p_value})")




```

```python
import pandas as pd
from scipy.stats import rankdata

def cliffs_delta(lst1, lst2):
    """Function to calculate Cliff's Delta"""
    m, n = len(lst1), len(lst2)
    lst1 = sorted(lst1)
    lst2 = sorted(lst2)
    j, k = 0, 0
    res = 0
    while j < m and k < n:
        if lst1[j] <= lst2[k]:
            j += 1
        elif lst1[j] > lst2[k]:
            res += m - j
            k += 1
    return res / (m * n)



superhost = df[df['host_is_superhost'] == 1]
non_superhost = df[df['host_is_superhost'] == 1]

# Loop through columns and calculate Cliff's Delta
for col in df.columns:
    if col == 'host_is_superhost':
        continue

    delta = cliffs_delta(superhost[col], non_superhost[col])
    if delta > 0.147:
        print(f"Column {col} has a large effect size (Cliff's Delta = {delta})")
    elif delta > 0.33:
        print(f"Column {col} has a very large effect size (Cliff's Delta = {delta})")
    elif delta > 0.474:
        print(f"Column {col} has an extremely large effect size (Cliff's Delta = {delta})")
```

```python
def test_significance(col):
    if col.name == 'host_is_superhost':
        return None
    p_value = mannwhitneyu(superhost[col.name], non_superhost[col.name]).pvalue
    if p_value < 0.05:
        return f"Column {col.name} is significantly different between Superhost and non-Superhost (p-value = {p_value})"
    else:
        return None

results = df.apply(test_significance)
results.dropna(inplace=True)
print(results)
```
